```python
import numpy as np
from matplotlib import pyplot as plt

import skimage.io
from skimage import color
from skimage.util import view_as_blocks
```

```python
testImage = skimage.io.imread("./3B_900k.jpg", as_gray=False)
testImage = color.rgb2gray(testImage)

plt.imshow(testImage, cmap="gray")
plt.xticks([])
plt.yticks([])
```

```python
#creating block
block_size = (1,1)
```

```python
#making blocks
test_image_blocks = view_as_blocks(testImage,block_size)
test_image_blocks.shape
```

```python
#flattening the blocks to make them in the same rown -1 is used to combine the last two indices of the block into a 1D array
flattened_blocks = test_image_blocks.reshape(test_image_blocks.shape[0],test_image_blocks.shape[1],-1)
```

```python
#getting the mean of the blocks
mean_blocks = np.mean(flattened_blocks, axis= 2)
plt.imshow(mean_blocks, cmap="gray",interpolation="nearest")
plt.xticks([])
plt.yticks([])
```
