```python
import numpy as np
from matplotlib import pyplot as plt

import skimage.io
from skimage import color
from skimage import measure
from skimage.util import view_as_blocks
```

```python
#importing image
testImage = skimage.io.imread("./3B_900k.jpg", as_gray=False)
plt.imshow(testImage, cmap="gray",interpolation="nearest")
plt.xticks([])
plt.yticks([])
```

```python
testImageGray = color.rgb2gray(testImage)
```

```python
#finding contours
contours = measure.find_contours(testImageGray, 0.5)
len(contours)
```

```python
#plotting contours
fig, axes = plt.subplots(1,2, figsize =(16,12),sharex=True,sharey=True)

ax = axes.ravel()

ax[0].imshow(testImageGray,cmap="gray",interpolation="nearest")
for n,contour in enumerate(contours):
    ax[0].plot(contour[:,1],contour[:,0])
plt.show()
```
