```python
import skimage
from skimage import io
from skimage import data, color
from skimage.filters import roberts,sobel

import numpy as np
import matplotlib.pyplot as plt

#for canny 
from skimage.feature import canny
from skimage.draw import polygon
```

```python
#importing image
testImage = skimage.io.imread("./3B_900k.jpg", as_gray=True)
plt.imshow(testImage ,cmap="gray", interpolation="nearest")
plt.xticks([])
plt.yticks([])
```

```python
#roberts edge detector
testImage_edge_roberts = roberts(testImage)
plt.imshow(testImage_edge_roberts ,cmap="gray", interpolation="nearest")
plt.xticks([])
plt.yticks([])
```

```python
#sobels edge detector
testImage_edge_sobel = sobel(testImage)
plt.imshow(testImage_edge_sobel ,cmap="gray", interpolation="nearest")
plt.xticks([])
plt.yticks([])
```

```python
#canny edge detection
testImage_edge_canny = canny(testImage,sigma= 1.21)
plt.imshow(testImage_edge_canny ,cmap="gray", interpolation="nearest")
plt.xticks([])
plt.yticks([])
```
